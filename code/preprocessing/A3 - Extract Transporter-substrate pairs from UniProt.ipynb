```python
start_rhea = "Xref=Rhea:"

def get_reaction_info(activity):
    reaction = activity.split(";")[0]
    reaction = reaction.replace("Reaction=", "")
    
    start_pos = activity.find(start_rhea)
    if start_pos != -1:
        rhea_ID = activity[start_pos + len(start_rhea):]
        rhea_ID = rhea_ID.split(",")[0]
    else:
        rhea_ID = np.nan
        
    return(rhea_ID, reaction)


def get_activtities(activities):
    activities = activities.split("CATALYTIC ACTIVITY: ")
    return(activities[1:])
```

```python
# Initialize an empty DataFrame
df_RHEA = pd.DataFrame(columns=["RHEA ID", "CHEBI_ID_list"])
rows = []  # Initialize an empty list to collect rows

# File path
file_path = join("..", "..", "data", "reactions", "rhea-reactions.txt")

# Read the file
with open(file_path, 'r') as file:
    Lines = file.readlines()

# Function to extract RHEA ID and CHEBI IDs


def extract_RHEA_ID_and_CHEBI_IDs(entry):
    rhea_id = None
    chebi_ids = []
    for line in entry:
        if line.startswith('ENTRY'):
            rhea_id = line.split()[1]
        elif line.startswith('EQUATION'):
            chebi_ids = [part.strip() for part in line.split()[1].split('+')]
    return rhea_id, chebi_ids


# Parse the file
start = 0
while start < len(Lines):
    try:
        end = Lines.index('///\n', start)
        entry = Lines[start:end]
        RHEA_ID, CHEBI_IDs = extract_RHEA_ID_and_CHEBI_IDs(entry)
        if RHEA_ID and CHEBI_IDs:
            rows.append({"RHEA ID": RHEA_ID, "CHEBI_ID_list": CHEBI_IDs})
        start = end + 1
    except ValueError:
        break

# Convert the list of rows to a DataFrame
df_new_rows = pd.DataFrame(rows)

# Concatenate the new rows DataFrame to the existing DataFrame
df_RHEA = pd.concat([df_RHEA, df_new_rows], ignore_index=True)

df_RHEA["RHEA ID"] = [float(ID.split(":")[-1]) for ID in df_RHEA["RHEA ID"]]
df_RHEA.to_pickle(
    join("..", "..", "data", "reactions", "RHEA_reaction_df.pkl"))
df_RHEA

# Display the DataFrame
print(df_RHEA)
```
